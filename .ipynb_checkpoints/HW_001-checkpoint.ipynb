```python
# imports
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
import datetime
#pd.set_option('display.max_rows', None)

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

import seaborn as sns
import matplotlib.pyplot as plt

# А заодно установим стиль графиков, что бы улучшить их восприятие:
sns.set(style="ticks")

import warnings
warnings.filterwarnings("ignore")
```

```python
import sklearn
sklearn.show_versions()
```

```python
# чтение и обработка данных
colls = ['timestamp','date']
def read_data(path: str, colls: list = colls) -> pd.DataFrame:
    df = pd.read_csv(path, parse_dates=[colls[0]])
    # раскручиваем дату
    df['hour']=df[colls[0]].dt.hour # час
    df['min'] = df[colls[0]].dt.minute #  
    df['day']=df[colls[0]].dt.day  # число месяца
    df['month'] = df[colls[0]].dt.month
    df['dayofweek']=df[colls[0]].dt.weekday  # день недели
    df[colls[1]]= df[colls[0]].dt.date # дата
     
    return df.drop([colls[0]], axis = 1)

test = read_data('test.csv')
train = read_data('train.csv')
train.head(3)
```

### Q1. Постройте логистическую регрессию,
- оптимизатор liblinear,
- остальные параметры дефолтные, на 6 признаках: 
- - ['gate_id', 'hour', 'min', 'day', 'month','dayofweek']
- Признаки делаем из "ts", в частности "min" - это минуты из времени. Используем масштабирование StandardScaler.

  
Каково будет качество на обучающей выборке и на валидации (округлите до целых и выберите ближайший ответ). 10 баллов.

```python
# разобъём трайн на две выборки в качестве val выбираем ноябрь- декабрь
limit = datetime.datetime.strptime('2022-11-01', '%Y-%m-%d').date()
train_train = train[train['date']<limit]
train_val = train[train['date']>=limit]
# выделяем X y
y_tr = train_train['user_id']
y_val = train_val['user_id']
X_tr = train_train.drop(['row_id', 'user_id', 'date'], axis = 1)
X_val = train_val.drop(['row_id', 'user_id', 'date'], axis = 1)
print (X_tr.shape)
X_tr.head()
```

```python
scaler = StandardScaler()
X_tr = scaler.fit_transform(X_tr)
X_val = scaler.transform(X_val)

clf = LogisticRegression( solver='liblinear')

clf.fit(X_tr, y_tr)
print(f'train accuracy {accuracy_score(y_tr, clf.predict(X_tr ))}') 
print(f'val accuracy {accuracy_score(y_val, clf.predict(X_val ))}') 
```

### Q3. Отметьте все правильные ответы для регрессии из Q2. 10 баллов

```python
# Регуляризация "l1" помогает при некоторых оптимизаторах - нет
solvers =  ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
for solver in solvers:
    print('\nsolver  ', solver)
    try:
        clf = LogisticRegression(random_state=1, solver=solver, penalty='l1')
        clf.fit(X_tr, y_tr)
        print(f'with l1 val accuracy {accuracy_score(y_val, clf.predict(X_val ))}')
        clf = LogisticRegression(random_state=1, solver=solver)
        clf.fit(X_tr, y_tr)
        print(f'without l1 val accuracy {accuracy_score(y_val, clf.predict(X_val ))}')
    except:
        print('С этим оптимизатором L1 не работает.')
```

```python
# "Multinomial" подход не хуже, а часто лучше "One-vs-rest ? " иногда да
solvers = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
multi_classes = [ 'multinomial', 'ovr']
for solver in solvers:
    print('\nsolver  ', solver)
    try:
        for multi_class in multi_classes:
            print('\nmulti_class ', multi_class)
            clf = LogisticRegression(random_state=1, solver=solver, multi_class=multi_class)
            clf.fit(X_tr, y_tr)
            print(f'\tval accuracy {accuracy_score(y_val, clf.predict(X_val ))}')
    except:
        print('С этим оптимизатором multinomial не работает.')
```

## Q10. Проверьте, помогает ли качеству регрессии из Q6 и оптимизатора saga больше данных
оцените разницу между 60%, 80% и 100% от train.

```python
train_dum=pd.get_dummies(train, columns=['gate_id','dayofweek', 'month', 'hour'], drop_first=False)
print (train_dum.columns)
train_dum.head()
```

```python
# разобъём трайн на две выборки в качестве val ввыбираем ноябрь- декабрь 
limit = datetime.datetime.strptime('2022-11-01', '%Y-%m-%d').date()
train_train = train_dum[train['date']<limit]
train_val = train_dum[train['date']>=limit]
# выделяем X y
y_tr = train_train['user_id']
y_val = train_val['user_id']
X_tr = train_train.drop(['row_id', 'user_id', 'date'], axis = 1)
X_val = train_val.drop(['row_id', 'user_id', 'date'], axis = 1)


scaler = StandardScaler()

X_tr = scaler.fit_transform(X_tr)
X_val = scaler.transform(X_val)
```

```python
for percent in [0.6, 0.8 ,1]:    
    ind = np.random.choice(X_tr.shape[0], int(X_tr.shape[0]*percent), replace=False)
    X_tr_new = X_tr[ind,:]
    y_tr_new = y_tr[ind]
    clf = LogisticRegression(random_state=1, solver='saga')
    clf.fit(X_tr_new, y_tr_new)
    print("\npercent ", percent)
    print(f'\t train  accuracy {accuracy_score(y_tr_new, clf.predict(X_tr_new ))}')
    print(f'\t val  accuracy {accuracy_score(y_val, clf.predict(X_val ))}')
    
```
